$$\int{(e^x \sin x + e^x \cos x) dx = e^x \sin x + c}$$

````python
# 不定积分，省略待定系数
integrate(exp(x)*sin(x) + exp(x)*cos(x), x)
````

## 求解常微分方程(ODE)

求解方程

$$y''-y=e^x$$

这是常系数线性微分方程，特征方程为

$$\lambda^2 - 1 = 0$$

特征根为 $\lambda = \pm 1$，齐次方程的通解为

$$\bar{y}(x) = c_1 e^x + c_2 e^{-x}$$

右端项为 $e^x$，采用待定系数法，可求得其一个特解

$$y^*(x) = \dfrac{x}{2}e^x$$

最后，非齐次方程的通解为

$$y(x) = c_1 e^x + c_2 e^{-x} + \dfrac{x}{2}e^x = c_2 e^{-x} + \left(c_1 + \dfrac{x}{2}\right)e^x$$


````python
y = Function('y')
dsolve(Eq(y(x).diff(x, x) - y(x), exp(x)), y(x))
````

### 再来一个求解微分方程例子

$$f''(x) - 2f'(x) +f(x) = sin(x)$$

````python
f, g = symbols('f g', cls=Function)
diffeq = Eq(f(x).diff(x, x) - 2*f(x).diff(x) + f(x), sin(x))
dsolve(diffeq, f(x))
````

## 求解特征值与特征矢量问题

已知矩阵

$$\boldsymbol{A} = 
\left[
\begin{array}{cc}
1 & 2\\
2 & 2
\end{array}
\right]$$

求其特征值 $\lambda$，满足

$$\boldsymbol{Ax}=\lambda \boldsymbol{x}$$

特征方程为

$$
\left|
\begin{array}{cc}
1-\lambda & 2\\
2       & 2-\lambda
\end{array}
\right|
= \lambda^2 - 3\lambda -2 = 0
$$

求得二根为

$$\lambda = \dfrac{-b\pm\sqrt{b^2 - 4ac}}{2a} = \dfrac{3\pm \sqrt{17}}{2}$$

````python
# 求特征值
Matrix([[1, 2], [2, 2]]).eigenvals()
````

````python
# 求特征矢量
Matrix([[1, 2], [2, 2]]).eigenvects()
````

## 输出Latex格式的公式表示

$$\int_{0}^{\pi} \cos^{2} x dx$$

````python
latex(Integral(cos(x)**2, (x, 0, pi)))
````

### 上式带有转义字符 \\\\

通过 print 命令输出后，可用于编辑 Latex 公式。

````python
print('\\int_{0}^{\\pi} \\cos^{2}{\\left (x \\right )}\\, dx')
````

## 将上述输出语句，置于 \$\$ $\cdot$ \$\$ 界限符内(MarkDown格)

```
$$\int_{0}^{\pi} \cos^{2}{\left (x \right )}\, dx$$
```

即生成数学式

$$\int_{0}^{\pi} \cos^{2}{\left (x \right )}\, dx$$

## 绘制图形

- 二维图
- 三维图

### 绘制二维图形

````python
# 绘制单曲线
from sympy import symbols
from sympy.plotting import plot
plot(x**2, (x, -5, 5))
````

````python
# 曲线迭加
from sympy import symbols
from sympy.plotting import plot
x = symbols('x')
p1 = plot(x*x)
p2 = plot(x)
p1.append(p2[0])
p1.show()
````

````python
# 绘制多曲线
plot((x**2, (x, -6, 6)), (x, (x, -5, 5)))
````

````python
# 绘制参数图
from sympy import symbols, cos, sin
from sympy.plotting import plot_parametric
u = symbols('u')
plot_parametric(cos(u), sin(u), (u, -5, 5))
````

````python
# 多参数图
plot_parametric((cos(u), sin(u)), (u, cos(u)))
````

### 绘制三维图形

````python
# 单曲面图
from sympy import symbols
from sympy.plotting import plot3d
x, y = symbols('x y')
plot3d(x*y, (x, -5, 5), (y, -5, 5))
````

````python
# 多曲面图
plot3d(x*y, -x*y, (x, -5, 5), (y, -5, 5))
````

````python
# 多曲面图，定义域不同
plot3d((x**2 + y**2, (x, -5, 5), (y, -5, 5)), (x*y, (x, -3, 3), (y, -3, 3)))
````

````python
# 三维曲线
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_line
u = symbols('u')
plot3d_parametric_line(cos(u), sin(u), u*u*u, (u, -25, 25))
````

````python
# 多曲线图
plot3d_parametric_line((cos(u), sin(u), u, (u, -25, 25)), (sin(u), u**2, u, (u, -25, 25)))
````

````python
# 参数曲面
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_surface
u, v = symbols('u v')
plot3d_parametric_surface(cos(u + v), sin(u - v), u - v, (u, -5, 5), (v, -5, 5))
````

````python
from sympy import *
from sympy.geometry import *
x = Point(0, 0)
y = Point(1, 1)
z = Point(2, 2)
zp = Point(1, 0)
````

````python
Point.is_collinear(x, y, z)
````

````python
Point.is_collinear(x, y, zp)
````

````python
t = Triangle(zp, y, x)
t.area
````

````python
t.medians[x]
````

````python
m = t.medians
intersection(m[x], m[y], m[zp])
````

````python
c = Circle(x, 5)
l = Line(Point(5, -5), Point(5, 5))
c.is_tangent(l) # is l tangent to c?
````

````python
l = Line(x, y)
c.is_tangent(l) # is l tangent to c?
````

````python
intersection(c, l)
````
